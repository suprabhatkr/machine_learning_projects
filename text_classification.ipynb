```python
import nltk
import random
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
```

```python
documents[:10]
```

```python
random.shuffle(documents)
print(len(documents))
```

```python
from nltk.corpus import stopwords
stop_words = list(set(stopwords.words('english')))
all_words = []
for w in movie_reviews.words():
    if w.lower() not in stop_words and len(w.lower())!=1:
        all_words.append(w.lower())
all_words = nltk.FreqDist(all_words)
```

```python
print(all_words.most_common(15))
print(all_words["stupid"])
```

```python
dataSet={}
pos_words=[]
neg_words=[]
for d in documents:
    if d[1]=='neg':
        for w in d[0]:
            if w.lower() not in stop_words and len(w.lower())!=1:
                neg_words.append(w.lower())
    else:
        for w in d[0]:
            if w.lower() not in stop_words and len(w.lower())!=1:
                pos_words.append(w.lower())
```

```python
pos_words=nltk.FreqDist(pos_words)
neg_words=nltk.FreqDist(neg_words)
print(pos_words.most_common(15))
print(neg_words.most_common(15))
```

```python
print("for stop type 'quit':")
inp=input()
while(inp!='quit'):
    posProb=1
    negProb=1
    listStr=inp.split(' ')
    for w in listStr:
        if w in all_words.keys():
            posProb*=(pos_words[w]/all_words[w])
            negProb*=(neg_words[w]/all_words[w])
    if posProb>2*negProb:
        print("Very Good")
    elif posProb>0.8*negProb:
        print("Good")   
    elif posProb*2<negProb:
        print("very Bad")
    else:
        print("Bad")
    inp=input()
```
